```python
import spacy
from spacytextblob.spacytextblob import SpacyTextBlob
import numpy as np
import pandas as pd
```

```python
nlp = spacy.load('en_core_web_md')
spacy_text_blob = SpacyTextBlob()
nlp.add_pipe(spacy_text_blob)
print(nlp.pipe_names)
```

```python
NLPdftest = pd.read_csv('DataForNLP2.csv')
```

```python
Q37list = NLPdftest["Q37"].tolist()
length = len(Q37list)
```

```python
for i in range(length):
    doc = nlp(Q37list[i])
    print('Polarity:', doc._.sentiment.polarity)
```

```python
for i in range(length):
    doc = nlp(Q37list[i])
    print('Sujectivity:', doc._.sentiment.subjectivity)
```

```python
for i in range(length):
    doc = nlp(Q37list[i])
    print('Assessments:', doc._.sentiment.assessments)
```

```python
for span in doc.sents:
    print(span.text, span._.sentiment.polarity, span._. sentiment.subjectivity)
```

```python
Sentiment = NLPdftest["Polarity"]
Subjectivity = NLPdftest["Subjectivity"]
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8,8))
plt.xlabel('Subjectivity')
plt.ylabel('Sentiment')
plt.title('Sentiment vs Subjectivity')
plt.scatter(Subjectivity, Sentiment, marker="+", c="b")
plt.savefig('sentvssub.png')
```
