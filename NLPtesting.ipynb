```python
from sklearn.neighbors import KernelDensity
def kde_sklearn(data, bandwidth = 1.0, kernel="linear"):
    kde_skl = KernelDensity(bandwidth = bandwidth, kernel=kernel)
    kde_skl.fit(data[:, np.newaxis])
    log_pdf = kde_skl.score_samples(xplot[:, np.newaxis]) # sklearn returns log(density)

    return np.exp(log_pdf)

plt.figure(figsize=(20,10))
xplot = np.linspace(-1,1,100)
PDFtophat = kde_sklearn(Polarity,bandwidth=0.01,kernel="tophat") #Complete
plt.plot(xplot,PDFtophat)
```

```python
PDFgaussian = kde_sklearn(Polarity,bandwidth=0.25,kernel="gaussian") #Complete (this is y)
PDFepanechnikov = kde_sklearn(Polarity,bandwidth=0.5,kernel="epanechnikov") #Complete (again y)
plt.figure(figsize=(20,10))
plt.plot(xplot,PDFgaussian,label="gaussian") #Complete
plt.plot(xplot,PDFepanechnikov,label="epanechnikov") #Complete
plt.legend(loc="best") #Complete
```
