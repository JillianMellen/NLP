```python
from sklearn.neighbors import KernelDensity
def kde_sklearn(data, bandwidth = 1.0, kernel="linear"):
    kde_skl = KernelDensity(bandwidth = bandwidth, kernel=kernel)
    kde_skl.fit(data[:, np.newaxis])
    log_pdf = kde_skl.score_samples(xplot[:, np.newaxis]) # sklearn returns log(density)

    return np.exp(log_pdf)

plt.figure(figsize=(15,10))
xplot = np.linspace(-1,1,100)
PDFtophat = kde_sklearn(Polarity,bandwidth=0.01,kernel="tophat") #Complete
plt.title('Kernel Density Histogram of Sentiment Scores')
#plt.ylabel('ystuff')
plt.xlabel('Sentiment')
plt.plot(xplot,PDFtophat)
plt.savefig('PolarityHistogram.png')
```

```python
PDFgaussian = kde_sklearn(Polarity,bandwidth=0.25,kernel="gaussian") 
#PDFepanechnikov = kde_sklearn(Polarity,bandwidth=0.5,kernel="epanechnikov") 
plt.figure(figsize=(15,10))
plt.plot(xplot,PDFgaussian,label="gaussian") 
#plt.plot(xplot,PDFepanechnikov,label="epanechnikov") 
plt.xlabel('Sentiment')
plt.title('Sentiment Scores Gaussian')
#plt.legend(loc="best") 
plt.savefig('Gaussians.png')
```
